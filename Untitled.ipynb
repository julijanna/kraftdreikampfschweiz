```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile

pd.set_option('display.max_colwidth', -1)

# read in excel file

clubs_df = pd.read_excel('Mitgliedervereine Adressen.xlsx', skiprows=2)

# fill NaNs with empty strings

clubs_df.fillna('', inplace=True)
```

```python
clubs_df['Adresse'] = clubs_df['Adresse'].str.replace('\n', '<br/>', regex=False)
```

```python
clubs_df.to_html(index=False, buf='clubs.html', escape=False)
```
